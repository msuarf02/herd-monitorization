```python
from ultralytics import YOLO

root = "/content/drive/MyDrive/TFG/animal-detection/"
dataset = "aerial_sheep_2"
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = YOLO('yolov8m.pt')
results = model.train(
   data= root + 'datasets/' + dataset +'/data.yaml',
   imgsz=640,
   epochs=20,
   batch=16,
   name=root + 'YOLOv8/'+ dataset +'/yolov8m'
)
```

```python
model = YOLO(root +'YOLOv8/' + dataset +'/yolov8m/weights/best.pt')
test_results = model.val(
    data = root + 'datasets/' + dataset +'/data.yaml',
    imgsz=640,
    batch=32,
    name= root+ 'YOLOv8/'+ dataset +'/test'
)
```
